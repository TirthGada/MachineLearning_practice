```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# Parameters for the normal distribution
mu = 0
sigma = np.sqrt(25)

# Function to integrate
def fX(x):
    return 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Calculate CDF using error function
def cdf(x):
    return 0.5 * (1 + erf((x - mu) / (sigma * np.sqrt(2))))

# Range of x values
x_values = np.linspace(-100, 100, 1000)

# Calculate numerical integration results and CDF values
numerical_results = [np.trapz([fX(xi) for xi in x_values[:i]], x_values[:i]) for i in range(len(x_values))]
cdf_values = [cdf(xi) for xi in x_values]

# Plot numerical results and CDF
plt.plot(x_values, numerical_results, label='Numerical Integration')
plt.plot(x_values, cdf_values, label='CDF (using erf)')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Numerical Integration vs. CDF')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters for the exponential distribution
lambda_ = 5

# (a) Generate random variates
random_variates = expon.rvs(scale=1/lambda_, size=10)

# (b) Calculate moments, median, and standard deviation
mean = expon.mean(scale=1/lambda_)
variance = expon.var(scale=1/lambda_)
skewness = expon.stats(scale=1/lambda_, moments='s')
kurtosis = expon.stats(scale=1/lambda_, moments='k')
median = expon.median(scale=1/lambda_)
std_deviation = expon.std(scale=1/lambda_)

# (c) Plot PDF and CDF
x_values = np.linspace(0, 10, 1000)
pdf = expon.pdf(x_values, scale=1/lambda_)
cdf = expon.cdf(x_values, scale=1/lambda_)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x_values, pdf)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution PDF')

plt.subplot(1, 2, 2)
plt.plot(x_values, cdf)
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Exponential Distribution CDF')

plt.tight_layout()
plt.show()

# Print calculated statistics
print("First Four Moments:")
print("Mean:", mean)
print("Variance:", variance)
print("Skewness:", skewness)
print("Excess Kurtosis:", kurtosis)
print("Median:", median)
print("Standard Deviation:", std_deviation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_loss_gain(entry_fee):
    num_trials = 10
    outcomes = np.random.choice(['H', 'T'], size=num_trials, replace=True, p=[0.5, 0.5])
    heads_count = np.count_nonzero(outcomes == 'H')
    
    gain = heads_count * 2
    loss = (num_trials - heads_count)
    
    net_profit = - gain + loss + entry_fee
    return net_profit

entry_fee_range = np.arange(5, 16, 1)
results = [calculate_loss_gain(entry_fee) for entry_fee in entry_fee_range]

plt.plot(entry_fee_range, results, marker='o')
plt.xlabel('Entry Fee (Rs)')
plt.ylabel('Loss/Gain (Rs)')
plt.title('Loss/Gain vs. Entry Fee')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def buffon_needle_experiment(num_drops, needle_length, line_spacing):
    count_crossed_lines = 0
    
    for _ in range(num_drops):
        mid_point = np.random.uniform(0, line_spacing/2)
        angle = np.random.uniform(0, np.pi/2)
        
        end_point_x = mid_point + (needle_length/2) * np.cos(angle)
        
        if end_point_x >= line_spacing / 2:
            count_crossed_lines += 1
    
    pi_estimate = (2 * needle_length * num_drops) / (line_spacing * count_crossed_lines)
    error = np.abs(np.pi - pi_estimate)
    
    return pi_estimate, error

num_drops_range = np.arange(1000, 10001, 1000)
needle_length = 1.0
line_spacing = 2.0

pi_estimates = []
errors = []

for num_drops in num_drops_range:
    pi_estimate, error = buffon_needle_experiment(num_drops, needle_length, line_spacing)
    pi_estimates.append(pi_estimate)
    errors.append(error)

# Plot the error as a function of the number of needle throws
plt.plot(num_drops_range, errors, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Needle Throws')
plt.ylabel('Error')
plt.title("Error in Buffon's Needle Experiment")
plt.grid()
plt.show()

# Print the estimated values of pi and errors
for i in range(len(num_drops_range)):
    print(f"Number of Needle Throws: {num_drops_range[i]}, Estimated : {pi_estimates[i]}, Error: {errors[i]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cdf_Z(z):
    if z < 0:
        return 0
    elif z < 1:
        return z**2 / 2
    elif z <= 2:
        return 1 - (2 - z)**2 / 2
    else:
        return 1

def pdf_Z(z):
    if z < 0:
        return 0
    elif z < 1:
        return z
    elif z <= 2:
        return 2 - z
    else:
        return 0

z_values = np.linspace(0, 2, 1000)
cdf_values = [cdf_Z(z) for z in z_values]
pdf_values = [pdf_Z(z) for z in z_values]

# Plot the cumulative distribution function
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(z_values, cdf_values)
plt.xlabel('z')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function for Z')

# Plot the probability density function
plt.subplot(1, 2, 2)
plt.plot(z_values, pdf_values)
plt.xlabel('z')
plt.ylabel('Probability Density')
plt.title('Probability Density Function for Z')

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

def calculate_loss_gain(entry_fee):
    n = 10
    p = 0.5
    mean_gain = 2
    mean_loss = -1
    
    expected_value = n * (p * mean_gain + (1 - p) * mean_loss) - entry_fee
    return expected_value

entry_fee_amounts = np.arange(5, 16)

loss_gain_values = [calculate_loss_gain(entry_fee) for entry_fee in entry_fee_amounts]

plt.plot(entry_fee_amounts, loss_gain_values, marker='o')
plt.xlabel('Entry Fee (Rs.)')
plt.ylabel('Loss/Gain (Rs.)')
plt.title('Loss/Gain vs. Entry Fee')
plt.grid()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

needle_length = 1.0
line_distance = 2.0
num_needle_throws = 100000

num_crossings = 0
for _ in range(num_needle_throws):
    center_y = np.random.uniform(0, line_distance / 2)
    angle = np.random.uniform(0, np.pi / 2)
    
    if center_y - (needle_length / 2) * np.sin(angle) < 0:
        num_crossings += 1

estimated_pi = 2 * (needle_length / line_distance) * (num_needle_throws / num_crossings)
error = np.abs(np.pi - estimated_pi)

print("Estimated :", estimated_pi)
print("Error:", error)

plt.figure(figsize=(8, 6))
plt.plot([0, num_needle_throws], [np.pi, np.pi], color='red', label='True ')
plt.plot([0, num_needle_throws], [estimated_pi, estimated_pi], color='blue', label='Estimated ')
plt.xlabel('Number of Needle Throws')
plt.ylabel('Value of ')
plt.title("Estimation of  using Buffon's Needle Experiment")
plt.legend()
plt.grid()
plt.show()
```
